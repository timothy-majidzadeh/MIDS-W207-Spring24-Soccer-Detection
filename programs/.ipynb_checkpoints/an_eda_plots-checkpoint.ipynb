```python
## import necessary libraries and packages
import pandas
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import math
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from google.colab import drive
drive.mount('/content/gdrive/', force_remount=True)
#data_file1 = "/content/gdrive/MyDrive/Colab Notebooks/topv_annotated"
data_file2 = "/content/gdrive/MyDrive/Colab Notebooks/top_view_labels.csv"

# importing sys
import sys
```

```python
ball_cols = ['Unnamed: 0','ball','ball.1','ball.2','ball.3']


data_file = data_file2
df_classes = pandas.read_csv(data_file)
df_classes = df_classes.drop(labels = [0,1], axis = 0)
df_classes_ball = df_classes[ball_cols]
df_classes_ball.head()
```

```python
print(df_classes_ball['ball'].describe())
print(df_classes_ball['ball.1'].describe())
print(df_classes_ball['ball.2'].describe())
print(df_classes_ball['ball.3'].describe())
```

```python
## account for missing
print(pandas.isnull(df_classes_ball['ball']).sum())
print(pandas.isnull(df_classes_ball['ball.1']).sum())
print(pandas.isnull(df_classes_ball['ball.2']).sum())
print(pandas.isnull(df_classes_ball['ball.3']).sum())
```

```python
df_classes.shape
```

```python
df_classes_ball = df_classes_ball.dropna()
```

```python
# Check out the spread for each coordinate parameter

df_classes_ball['ball'] = df_classes_ball['ball'].astype(float)
df_classes_ball['ball.1'] = df_classes_ball['ball.1'].astype(float)
df_classes_ball['ball.2'] = df_classes_ball['ball.2'].astype(float)
df_classes_ball['ball.3'] = df_classes_ball['ball.3'].astype(float)

print(round(df_classes_ball['ball'].std(),1))
print(round(df_classes_ball['ball.1'].std(),1))
print(round(df_classes_ball['ball.2'].std(),1))
print(round(df_classes_ball['ball.3'].std(),1))
```

```python
# Renaming the frame coordinates
df_classes_ball.rename(columns = {'ball':'bb_height','ball.1':'bb_left','ball.2':'bb_top','ball.3':'bb_width'}, inplace = True)
df_classes_ball.head()
```

#The variability of the bounding box of the object may suggest a shift in the center of image. This shift makes a difference between choosing a Neural Network base model and a Convolutional Neural Network Model which does a better job predicting wheen the image moves.

```python
df_classes_ball.agg({'bb_left': ['count','mean','min','max','std'],'bb_top': ['count','mean','min','max','std'],'bb_width': ['count','mean','min','max','std'],'bb_height': ['count','mean','min','max','std']})

```

```python
#The variability of the bounding box of the object also implies a shift in the center of image.
df_classes_ball = df_classes_ball[['bb_left','bb_top','bb_height','bb_width']]
df_classes_ball.groupby(['bb_height','bb_width']).count().sort_values(by='bb_left', ascending=False)
```
